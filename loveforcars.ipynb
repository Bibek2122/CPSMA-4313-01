```python
# Bibek Thapa - I have chosen fastest cars
import requests
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO
```

```python
url = "https://en.wikipedia.org/wiki/List_of_fastest_production_cars_by_acceleration"
```

```python
headers = {"User-Agent": "Mozilla/5.0"}
r = requests.get(url, headers=headers)

html_contents = r.text
```

```python
html_soup = BeautifulSoup(html_contents, "lxml")
```

```python
tables = html_soup.find_all("table")

len(tables)
```

```python
df = pd.read_html(StringIO(str(tables)))[0]

df
```

```python
df.to_csv("fastest_cars_acceleration.csv", index=False)
```
