```python
from google.cloud import bigquery
import pydata_google_auth

credentials = pydata_google_auth.get_user_credentials(
    ['https://www.googleapis.com/auth/cloud-platform']
)

PROJECT_ID = "big-query-488604"

client = bigquery.Client(credentials=credentials, project=PROJECT_ID)

print("Authenticated Successfully using project:", PROJECT_ID)
```

```python
# Query to find the second most popular starting station
# First i did count trips per station
# Finally I used OFFSET 1 to get the second result

query_second_station = """
SELECT start_station_name, COUNT(*) AS number_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY number_trips DESC
LIMIT 1 OFFSET 1
"""

client.query(query_second_station).to_dataframe()
```

Result Explanation:
After grouping the dataset by starting station and counting the total number of trips, the stations were ordered in descending order based on trip frequency.The most popular starting station was "21st/Speedway @ PCL".The second most popular starting station was "21st & Speedway @PCL" with a total of 71,145 trips. This result was obtained by ordering the stations by total trips and selecting the second highest value.

```python
# Query to count round trips lasting more than 60 minutes
# A round trip is when start_station_name = end_station_name
#duration_minutes > 60 is also required

query_round_trips = """
SELECT COUNT(*) AS round_trips_over_hour
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
AND duration_minutes > 60
"""

client.query(query_round_trips).to_dataframe()
```

Result Explanation:
To answer this question, the dataset was filtered to include only trips, where the starting and ending stations were the same, indicating a round trip. Additionally, only trips lasting more than 60 minutes were included.After applying these conditions and counting the results, there were 70,666 trips that were round trips and lasted over one hour.
