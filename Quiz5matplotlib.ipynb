```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns


df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv')
```

```python
ax = sns.scatterplot(data = df, x = 'SepalLength', y = 'SepalWidth')
ax.set(title = "Length vs Width",
       xticks = [x for x in range(4,9,1)])
plt.show()
```

I could not get different colors in class because I did not include the hue='Class' parameter in the scatterplot. Since no grouping variable was provided, seaborn treated all the data as one group and used the default blue color for every point.

```python
color_map = {
    'Iris-setosa': 'red',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'blue'
}
for species, color in color_map.items():
    subset = df[df['Class'] == species]
    plt.scatter(subset['SepalLength'], subset['SepalWidth'], color=color, label=species)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Length vs Width")
plt.legend()  
plt.show()
```

In this graph, I used matplotlib to plot Sepal Length on the x-axis and Sepal Width on the y-axis. To show different colors for each flower class, I created a dictionary that assigns a specific color to each class (red for Iris-setosa, green for Iris-versicolor, and blue for Iris-virginica). I then used the .map() function to convert the class names into their corresponding colors and passed those colors into the scatter plot using the c= parameter.

In class, I could not get different colors because I did not specify how the classes should be colored. When no color mapping is provided, matplotlib (or seaborn without hue) uses one default color for all points, which is why the graph appeared all blue. We didn’t see colors because Matplotlib cannot use text labels as colors directly. To show colors, each class must be mapped to a specific color or plotted separately with labels. Unlike Seaborn, Matplotlib requires this extra step to display colored points and a legend.
