```python
#Bibek Thapa
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/refs/heads/main/Data_Sets/Activity_Dataset_V1.csv"
df = pd.read_csv(url)

df.head()
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm", 
    linewidths=0.5    
)
plt.title("Association Heatmap")
plt.show()
```

So this assignment explores associations between variables in an activity dataset. I calculated correlations to see how different health and activity measurements relate to each other. I visualized these relationships using a heatmap, which allows many variables to be viewed at once.
I used a color gradient coolwarm to make the relationships easier to understand. This makes patterns easier to identify. I used the correlation function with numeric_only=True which helps to include all numeric variables in the dataset. This ensured the heatmap displayed the maximum number of variables and their relationships.
