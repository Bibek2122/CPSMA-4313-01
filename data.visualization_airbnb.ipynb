```python
#Bibek Thapa
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airbnb = pd.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv")
```

```python
# Group by neighbourhood group and get max price including price
max_price = airbnb.groupby("neighbourhood_group")["price"].max()

# Create bar graph
plt.figure()
max_price.plot(kind="bar")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Maximum Price ($)")
plt.title("Maximum Airbnb Price by Neighbourhood Group")
plt.show()
```

```python
# Group by neighbourhood_group and room_type, then calculating mean price
avg_price = airbnb.groupby(
    ["neighbourhood_group", "room_type"]
)["price"].mean().unstack()

# Create multiple bar chart
plt.figure()
avg_price.plot(kind="bar")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price ($)")
plt.title("Average Airbnb Price by Neighbourhood Group and Room Type")
plt.legend(title="Room Type")
plt.show()
```
