```python
from google.cloud import bigquery
import pydata_google_auth

credentials = pydata_google_auth.get_user_credentials(
    ['https://www.googleapis.com/auth/cloud-platform']
)

PROJECT_ID = "big-query-488604"

client = bigquery.Client(credentials=credentials, project=PROJECT_ID)

print("Authenticated Successfully using project:", PROJECT_ID)
```

```python
query_preview = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
"""

client.query(query_preview).to_dataframe()
```

The output shows the first 5 rows of the Austin Bikeshare trips dataset. Each row represents a single bike trip, including details such as the trip ID, start and end times, bike ID, trip duration, start and end stations, and subscriber type.  This preview helps us quickly understand the structure of the table and the kind of information available for analysis.

```python
query_avg = """
SELECT 
    start_station_name,
    AVG(duration_minutes) AS avg_trip_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY avg_trip_time DESC
"""

df_avg = client.query(query_avg).to_dataframe()
df_avg.head()
```

The output shows the average trip duration (in minutes) for each start station in the Austin Bikeshare dataset. 
The table is sorted in descending order, so the stations with the longest average trips appear first.  For example, trips starting at "East 5th/Shady Ln" have the highest average duration (~197 minutes),  while other stations like "Lakeshore @ Austin Hostel" have shorter average trips (~65 minutes). This helps identify which stations tend to have longer or shorter rides on average.

```python
query_count = """
SELECT 
    start_station_name,
    COUNT(*) AS trip_count
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY trip_count DESC
"""

df_count = client.query(query_count).to_dataframe()
df_count.head()
```

The output shows the total number of trips starting from each station in the Austin Bikeshare dataset. 
The table is sorted in descending order, so the most frequently used stations appear first. 
For example, "21st/Speedway @ PCL" has the highest number of trips (108,559), indicating it is a very popular starting point, while other stations like "26th/Nueces" have fewer trips (44,179). 
 This helps identify which stations are the busiest in terms of trip volume.
