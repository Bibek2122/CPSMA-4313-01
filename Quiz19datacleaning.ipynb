```python
import pandas as pd

# Loading the dataset
url = 'https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv'
df = pd.read_csv(url)

# Previewing the first few rows
df.head()
```

```python

df.isnull().sum()
```

```python
# Filling missing review counts with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

```python
# Filling missing names with a placeholder string
df['name'] = df['name'].fillna("Unknown")
df['host_name'] = df['host_name'].fillna("Unknown")
```

```python
# Verifying target columns no longer have nulls
print("Remaining nulls per column:")
print(df.isnull().sum())
```

```python
# 1. Examining how many entries are null
print(df.isnull().sum())

# 2. 'reviews_per_month' with 0 (since null = no reviews)
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# 3. 'name' with a placeholder (since it's a text column)
df['name'] = df['name'].fillna("Unknown")

# 4. Showing the first few rows to confirm it worked
df.head() #Looking final talbe too
```

I performed a thorough data cleaning process on the NYC Airbnb dataset to ensure it was suitable for analysis. I began by loading the dataset into a Jupyter Notebook using pandas and used the isnull().sum() function to identify and quantify the missing entries across all columns. To address these gaps, I performed imputation on two key columns: I filled the missing values in reviews_per_month with 0, logically assuming that a lack of data indicated zero reviews, and I replaced null entries in the name column with the placeholder "Unknown". This allowed me to "patch" the dataset without deleting valuable rows, ensuring that future calculations and visualizations remain accurate and inclusive of all listings.
