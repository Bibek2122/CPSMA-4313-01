```python
import pdfplumber
import pandas as pd

all_data = []

with pdfplumber.open("sample-tables.pdf") as pdf:
    for page in pdf.pages:
        tables = page.extract_tables()
        for table in tables:
            # skip header for now, just collect all rows
            for row in table:
                all_data.append(row)
all_data[:5]
```

```python
max_cols = max(len(row) for row in all_data)
```

```python
cleaned_data = []

for row in all_data:
    # add empty strings to make all rows same length
    if len(row) < max_cols:
        row += [''] * (max_cols - len(row))
    cleaned_data.append(row)
```

```python
df_all = pd.DataFrame(cleaned_data)
df_all.head()
```

```python
df_all = df_all.applymap(lambda x: x.replace('\n',' ') if isinstance(x, str) else x)
```

```python
df_all.to_csv("output_clean.csv", index=False)
```
