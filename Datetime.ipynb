```python
import pandas as pd  # Import pandas for data analysis

# Loading dataset from URL
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/IOT-temp.csv"
df = pd.read_csv(url)

# Displaying first 5 rows to understand the data
df.head()
```

This cell loads the dataset and displays the first few rows to understand the data.

```python
df["noted_date"] = pd.to_datetime(df["noted_date"], format="%d-%m-%Y %H:%M")

df.info()
```

This cell converts the noted_date column into datetime format for time-based analysis.

```python
min_date = df["noted_date"].min()
max_date = df["noted_date"].max()

min_date, max_date
```

```python
df = df.sort_values("noted_date")

df["time_diff"] = df["noted_date"].diff()

largest_interval = df["time_diff"].max()

largest_interval
```

This cell calculates the time gaps between readings and finds the largest interval.

```python
avg_datetime = df["noted_date"].mean()

avg_datetime
```

This cell computes the average (mean) date and time of all records.

```python
# Swapping the month and day in the string
target_date = pd.to_datetime("2018-09-11").date()

 #Keeping string and specify the format
# target_date = pd.to_datetime("09-11-2018", dayfirst=True).date()

filtered = df[df["noted_date"].dt.date == target_date]

# Selecting only outdoor readings
filtered_outdoor = filtered[filtered["out/in"] == "Out"]

# Computing average temperature
avg_temp = filtered_outdoor["temp"].mean()
avg_temp
```

This cell filters data for November 9, 2018, and calculates the average outdoor temperature for that day.

```python
print("Minimum Date and Time:", min_date)
print("Maximum Date and Time:", max_date)
print("Largest Interval Between Checks:", largest_interval)
print("Average Datetime:", avg_datetime)
print("Average Outdoor Temperature on 09-11-2018:", avg_temp)
```
