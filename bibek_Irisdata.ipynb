```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv(
    "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv"
)
iris.head()
```

I will examine dataset based on the structure and summary statistics of the datasets.

```python
print(iris.columns)
```

```python
iris.describe()
```

```python
iris['Class'].value_counts().plot(kind='bar')
plt.title('Count of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(iris['PedalLength'], bins=20)
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(iris['PedalLength'])
plt.title('Box Plot of Petal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
sns.violinplot(y=iris['PedalLength'])
plt.title('Violin Plot of Petal Length')
plt.show()
```
